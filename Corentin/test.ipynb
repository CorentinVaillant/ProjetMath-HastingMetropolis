```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# Example Markov transition matrix
P = np.array([
    [0.2, 0.5, 0.3],
    [0.1, 0.6, 0.3],
    [0.4, 0.2, 0.4]
])

# Create directed graph
G = nx.DiGraph()

# Add edges with weights (transition probabilities)
num_states = P.shape[0]
for i in range(num_states):
    for j in range(num_states):
        if P[i, j] > 0:  # Only add edges with nonzero probability
            G.add_edge(i, j, weight=P[i, j])

# Position nodes using spring layout
pos = nx.spring_layout(G)

# Draw the graph
plt.figure(figsize=(6, 6))
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=700, arrowsize=20)

# Add edge labels (transition probabilities)
edge_labels = {(i, j): f'{P[i, j]:.2f}' for i, j in G.edges()}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=12)

plt.show()
```
